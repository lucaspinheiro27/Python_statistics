```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hf://datasets/hugginglearners/data-science-job-salaries/ds_salaries.csv")
df
```

```python
ax = df.boxplot(by='experience_level', column='salary_in_usd')
ax.set_xlabel('')
ax.set_ylabel('Salary in USD')
plt.suptitle('')
```

EN Entry-level / Junior MI Mid-level / Intermediate SE Senior-level / Expert EX Executive-level / Director
* Executive - level experts stands out as having the highest salaries compared to others, while entry level people have the lower salaries

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='experience_level', y='salary_in_usd', data=df)
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.title('Salary Distribution by Experience Level')
plt.show()
```

This violin plot demonstrates a concentration in distribution to a lesser extent categorized by experience level, this is not as obvious compared to boxplots.

```python
experience = ['SE', 'EX']
experience_zip = df.loc[df.experience_level.isin(experience)]

def hexbin(x, y, color, **kwargs):
  cmap = sns.light_palette(color, as_cmap=True)
  plt.hexbin(x, y, gridsize=25, cmap=cmap, **kwargs)

g = sns.FacetGrid(experience_zip, col='experience_level', col_wrap=2, height=6)
g.map(hexbin, 'salary_in_usd', 'remote_ratio', extent=[0, 600000, 0, 100])
g.set_axis_labels('Salary in USD', 'Remote ratio')
plt.show()
```

* Two particular variables have received higher salaries compared to others. By plotting these two variables in particular, the picture is much clearer, the salaries are indeed higher for executive level and seniors.
