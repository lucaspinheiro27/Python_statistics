# Student's t-Distribution

* Reference for a distribution of a sample mean, the difference from the two means, regression parameters
* a Family of distributions resembling the normal distribution but with thicker tails



$$
\bar{x} \pm t_{n-1}(0.05) \cdot \frac{s}{\sqrt{n}}
$$

```python
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import chi2, poisson
import numpy as np
from scipy.stats import expon
plt.style.use('dark_background')
```

* Array of 10 values that follow a t-distribution of 6 degrees of freedom

```python
t.rvs(df=6, size=10)
```

* Calculate th p-value that corresponds to the p-statistic with 10 degrees of freedom

```python
t.cdf(x=-1.5, df=10)
```

* Find two-tailed hyphotesis test to calcute the p-value

```python
(1 - t.cdf(x=2.14, df=20))*2
```

* Plotting a t-distribution

```python
x = t.rvs(df=12, size=10000)

plt.hist(x, density=True, edgecolor='black', bins=20)
```

* Using a density curve

```python
sns.kdeplot(x)
```

# Binomial Distribution

* Frequency distribution of the number of successes (x) in a givem number of trials(n) with specified probability (p)  of success in each trial

```python
stats.binom.pmf(2, n=5, p=0.1)
```

* determining the probability of x or fewer success in n trials

```python
stats.binom.cdf(2, n=5, p=0.1)
```

# Chi-Square Distribution

* Concerned with counts of subjects or items falling into catgories.

* The chi-square statistic measure the extent of departure from what you would expect in a null model.

```python
x = np.arange(0, 20, 0.001)
```

* 4 Degrees of Freedom

```python
plt.plot(x, chi2.pdf(x, df=4))
```

* plotting multiple Chi-Square Distributions

```python
plt.plot(x, chi2.pdf(x, df=4), label='df: 4')
plt.plot(x, chi2.pdf(x, df=8), label='df: 8')
plt.plot(x, chi2.pdf(x, df=8), label='df: 12')

plt.legend()
```

# Poisson Distribution

* For events that occur at a constant rate, the number of events per time unit of time or space.

$$
P(X = x) = \frac{\lambda^x e^{-\lambda}}{x!}
$$


  * : mean number of successes that occur during a specific interval
  * x: number of successes
  * e: a constant equal to approximately 2.71828


* mean value and sample size

```python
x = stats.poisson.rvs(2, size=100)
```

* Equal to some value

```python
poisson.pmf(k=5, mu=2)
```

* Less than some values

```python
poisson.cdf(k=4, mu=2)
```

* Greater tha some value

```python
1-poisson.cdf(k=20, mu=15)
```

```python
plt.hist(x, density=True, edgecolor='black', bins=7)
```

# Exponential Distribution

* Model the time or distance between one event and the next.

$$
f(x; \lambda) =
\begin{cases}
\lambda e^{-\lambda x} & \text{if } x \geq 0, \\
0 & \text{if } x < 0.
\end{cases}
$$

* Interest in calculating how long we have to wait until a certain event occurs.

* Calculate probability that x is less than 50 when mean rate is 40

```python
expon.cdf(x=50, scale=40)
```

* Generating an exponential distribution with sample size

```python
x = expon.rvs(scale=40, size=1000)
plt.hist(x, density=True, edgecolor='black')
```
