1)  You survey 20 shoppers to see what type of soft drink they like best, Brand A or
Brand B. The results are: A, A, B, B, B, B, B, B, A, A, A, B, A, A, A, A, B, B, A, A. Which
brand do the shoppers prefer? Make a frequency table and explain your answer.

```python
import pandas as pd

data = {'Shopper': range(1, 21),
        'Soft Drink Preference': ['A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'A', 'A', 'A', 'B', 'A', 'A', 'A', 'A', 'B', 'B', 'A', 'A']}

df = pd.DataFrame(data)

df
```

```python
frequency_table = df['Soft Drink Preference'].value_counts()
frequency_table['Total'] = frequency_table.sum()
frequency_table
```

* Eleven clients prefer Brand A over Brand B. The difference between the two brands is 2, which is relatively small. By creating a relative frequency table, we can determine that Brand A accounts for 55% of the total sample size of 100%. This method often makes it easier to interpret the results.

```python
percentage = (frequency_table / frequency_table['Total']) * 100
percentage
```

```python
difference = frequency_table['A'] - frequency_table['B']
proportion = difference / frequency_table['Total']
print(f"The difference beteween the two drinks is: {difference}")
print(f"The proportion of A-B is: {proportion}")
```

2) A local city government asks voters to vote on a tax levy for the local school district.
A total of 18,726 citizens vote on the issue. The Yes count comes in at 10,479, and the
rest of the voters say No.
a. Show the results in a frequency table.
b. Why is it important to include the total number at the bottom of a frequency table

```python
votes = {'Vote': ['Yes', 'No']}
votes = pd.Series([10479, 18726 - 10479], index=votes['Vote'])
votes['Total'] = votes.sum()
votes
```

* Frequencies are acceptable as long as the total number is included at the bottom of the frequency table. The total number of citizens mentioned is approximately 18,726. To determine the difference in the number of voters who said "No," we need to subtract the number of voters who said "Yes" from the total number of voters.

```python
percentage = (votes / votes['Total']) * 100
percentage
```

3)  A zoo asks 1,000 people whether they’ve been to the zoo in the last year. The pollsters
count the results and find that 592 say yes, 198 say no, and 210 don’t respond.
a. Show the results in a frequency table.
b. Explain why you need to include the people who don’t respond.

```python
poll_zoo = {'Went to the zoo': ['Yes', 'No', 'Without answer']}
poll_zoo = pd.Series([592, 198, 210], index=poll_zoo['Went to the zoo'])
poll_zoo['Total'] = poll_zoo.sum()
poll_zoo
```

* In this particular case, it is important to mention the number of people who did not respond. If this number is not included, the total will not add up to 100%, resulting in a biased outcome.

```python
percentages = (poll_zoo / poll_zoo['Total']) * 100
percentages
```

4)  Suppose you ask 1,000 people to identify from a list of five vacation spots which ones
they’ve already visited. The frequencies you receive are as follows: Disney World, 216;
New Orleans, 312; Las Vegas, 418; New York City, 359; and Washington, D.C., 188.
a. Explain why creating a traditional relative frequency table doesn’t make sense here.
b. How can you summarize this data with percents in a way that makes sense?

If you have only a frequency table, can you find the corresponding relative frequency
table? Conversely, if you have only a relative frequency table, can you find the corresponding frequency table? Explain your answer

```python
data = {'Vacation Spot': ['Disney World', 'New Orleans', 'Las Vegas', 'New York City', 'Washington, D.C.'],
        'Frequency': [216, 312, 418, 359, 188]}

df = pd.DataFrame(data)
df
```

```python
freq_total = df['Frequency'].sum()
print(f"Total frequency:{freq_total}")
```

```python
df['% Who have been there'] = df['Frequency'] / 1000
df['% Who havent been there'] = 1 - df['% Who have been there']
df
```

* The total does not sum to 1,000 people because individuals can choose between one or more categories; therefore, we cannot assume that each category corresponds to exactly one person. The relative frequencies represent the proportion of people in each category, which adds up to 100%. To find the percentage of people who have not participated in any category, we need to subtract the percentage of people present in the categories from the total count of 100%.
